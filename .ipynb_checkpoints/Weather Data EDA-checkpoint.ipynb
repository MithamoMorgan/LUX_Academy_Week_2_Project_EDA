## 2.Statisticsl Summanry

```python
# Statastical summary of the dataset
weather_data.describe().T
```

```python
# Using a box plot to identify columns with outliers

# Creating a new figure of size 10 * 8 inches
plt.figure(figsize = (10,8))
# Creating a box plot
sns.boxplot(data = weather_data)
# Display the chart
plt.show()
```

```python
# Printing rows with outliers using IQR method for each numeric column

def outlier_detection(weather_data):
    # Initialize an empty dict to hold outliers for each numeric feature
    outliers_dict = {}
    # Loop through each column and fetch the outliers if any
    for column in weather_data.select_dtypes(include=[float, int]).columns:
        # Calculate lower and upper bounds
        Q1 = weather_data[column].quantile(0.25)
        Q3 = weather_data[column].quantile(0.75)
        # Calculate interquatile range
        IQR = Q3 - Q1
        # Find the outliers
        outliers  = weather_data[(weather_data[column] < (Q1 - 1.5 * IQR)) | (weather_data[column] > (Q3 + 1.5 * IQR))]
        # Store the outliers in the dict with column name as key
        outliers_dict[column] = outliers
     # Return the dict    
    return outliers_dict
# Call the  outlier_detection function
outliers_dict = outlier_detection(weather_data)

# Loop through the dictionary and print outliers for each feature
for feature, outliers in outliers_dict.items():
    print(f"Outliers in {feature}:")
    print(outliers, "/n")
```
